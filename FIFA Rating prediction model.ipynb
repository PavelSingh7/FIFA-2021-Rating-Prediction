### Now lets handle datatypes some of the features

#### Height 

```python
def height_conversor(Height):
    feet, inches = Height.split("'")
    feet = int(feet)
    inches = int(inches.strip('"'))
    total_inches = feet * 12 + inches
    cm = total_inches * 2.54
    return cm
a['Height_cm']=a['Height'].apply(height_conversor)
```

## Weight

```python
def weight_conversor(Weight):
    weight_kgs=float(Weight.split('lbs')[0].strip())
    return weight_kgs * 0.45359237     
```

```python
a['Weight_kgs']=a['Weight'].apply(weight_conversor)
```

### Value,Wage,Release Clause

```python
def clean_cols(value):
    value=value.lstrip('€')
    suffix=value[-1]
    if suffix=='M':
        return float(value[:-1])*1000000
    elif suffix=='K':
        return float (value[:-1])*1000
    else:
        return float(value)
```

```python
a['Values_in_euros']=a['Value'].apply(clean_cols)
a['Wage_in_euros']=a['Wage'].apply(clean_cols)
a['Release_clause_euros']=a['Release Clause'].apply(clean_cols)
```

# Regression model to predict rating of players
###### As we are taking a lot of input variables we will be using Lasso Regeression as it bring most of the coefficient to zero and only takes important variables

```python
from sklearn.model_selection import train_test_split
```

```python
#defining independent variables
x= df.iloc[:,9:59]
```

```python
for col in x.columns:
    if x[col].isnull().any():
        x[col].fillna(x[col].mean(),inplace=True)
    print(x.head())
```

```python
x.isnull().sum()
```

```python
y=df.loc[:,['OVA']]
```

```python
x.head()
```

##### Dropping Height and Weight as it does not affect a players skills much

```python
x=x.drop(['Height_cm','Weight_kgs'],axis=1)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Standardizing the Scale


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
Las = Lasso(alpha=0.01)
```

```python
 Las.fit(X_train_scaled,y_train)
```

```python
prediction= Las.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,prediction)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print("Accuracy",np.mean(cross_val_score(Las,x,y,cv=5,scoring="r2")))
```

```python
# Print non-zero coefficients (selected features)
selected_features = x.columns[Las.coef_ != 0]
print('Selected Features:', selected_features)
```
